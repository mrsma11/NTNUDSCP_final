# DSCP Final Project

This code converts all the data with a format with multiple "lists in the list" into a two-column ASCII table. I save the original data in "Data_from_Catalog.py", and the final output of a two-column ASCII table is "ESSENCEm158.txt". The .txt file will be created after running this code. Also, you will see a plot of a spectrum of the supernova ESSENCEm158.

- Visit <a href="https://github.com/mrsma11/NTNUDSCP_final"> my Github page of this project</a>  for a complete description and all the files used in this code.
- Check if "Data_from_Catalog.py" is uploaded successfully before running this code. If not, you can download it <a href="https://github.com/mrsma11/NTNUDSCP_final/blob/main/Data_from_Catalog.py"> here</a>.

```python
'''
Example Codes 1: plot.py
'''

#load the modules and needed files
import matplotlib.pyplot as plt
import numpy as np 
import Data_from_Catalog

#Create a new .txt file to write in data (target: 2 columns ASCII )
wrt=open(r'ESSENCEm158.txt','w') 

#import the observational data, the format is similar to  Fig.2
data=[]
data=Data_from_Catalog.f() 

#For data visualization
wl=[] 
flux=[]

#Read data and rewrite in a new file (list in list --> 2 column ASCII)
#all "append" in this section will be useful for data visualization

for i in data:
    for j in i:
        if float(j)>1000: #basically it's wavelength, put it in the first column
            wl.append(float(j)) 
            wrt.write(j) 
            wrt.write(' ') #space 
        else:   
            flux.append(float(j)) 
            F=str(float(j)*1.e15)  #calibration to 1.e-15 for research purpose
            wrt.write(F)
            wrt.write('\n')
wrt.close()

#Data Visualization 
plt.plot(wl,flux)
plt.xlim(5500,7000)
plt.title("ESSENCEm158") 
plt.xlabel(r"rest wavelength ($\AA$)")
plt.ylabel(r"relative flux")
plt.show()
```

```python
'''
Example Codes 2: format conversion.py
'''

import json
import matplotlib.pyplot as plt

#open the files
with open(r'ESSENCEm158.json') as f:
    source=json.load(f) #source is a dict
    
#write data in terms of 2-column ASCII (like Fig. 1)
wrt=open(r'ESSENCEm158_from_json.txt','w')
    
#required data: only spectra, the other information are trivial in my research
spec=source['ESSENCEm158']['spectra'][0]['data']

#another way to arange these data 
for i in spec:
    wrt.write(i[0])
    wrt.write(' ')
    wrt.write(str(float(i[1])*1.e15))
    wrt.write('\n')
wrt.close()

print("The conversion has finished, please check the repository!")
```
